```python
import pandas as pd
import numpy as np
```

```python
dat = {
    'height':[150,160,170,180,190,200,210],    #x
    'weight':[60,65,70,75,80,85,90]            #y
}
```

```python
df = pd.DataFrame(dat)
```

```python
df.head()
```

```python
karl_corr = df['height'].corr(df['weight'])
karl_corr
#karl_corr = df['height'].corr(df['weight'])
#karl_corr
```

```python
spear_corr = df['height'].corr(df['weight'],method = 'spearman')
spear_corr
#spear_corr = df['height'].corr(['weight'],method = 'spearman')
#spear_corr
```

```python
df['rank1'] = df['height'].rank()
df['rank2'] = df['weight'].rank()
df.head()
# HERE RANKING WILL ALWAYS BE IN ASCENDING ORDER
```

```python
df['d'] = df['rank1'] - df['rank2']
df.head()
```

```python
df['d2'] = df['d'] ** 2
df
```

```python
sum =  df['d2'].sum()
sum
```

```python
n = len(df)
n
```

```python
R = 1-((6*sum)/(n**3-n))
R
```

```python
#calculate mean and standard deviation of height and weight column manually
```

```python
# MEAN
mean_height = df['height'].sum()/len(df)
print(mean_height)
mean_weight = df['weight'].sum()/len(df)
print(mean_weight)
```

```python
# STANDARD DEVIATION
std_height = np.sqrt(((df['height']-mean_height)**2).sum()/len(df))
print(std_height)
std_weight = np.sqrt(((df['weight']-mean_weight)**2).sum()/len(df))
print(std_weight)
```

```python
# karl_pearson manually
Xh=df['height'].sum()/len(df)
Yw=df['weight'].sum()/len(df)
df['X']=(df['height']-Xh)
df['Y']=(df['weight']-Yw)
x=df['X'].sum()
y=df['Y'].sum()
df['XY']=df['X']*df['Y']
a=df['XY'].sum()
df['x2']=df['X']**2
df['y2']=df['Y']**2
b=df['x2'].sum()
c=df['y2'].sum()
r=a/np.sqrt(b*c)
print(r)
```
