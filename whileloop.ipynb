```python
# wap to find sum of n natural numbers
n=int(input("enter the last number"))
i=1
sum=0
while (i <= n):
  sum = sum + i
  i =i+1
  print("the sum is:",sum)
print("the sum is:", sum)
```

```python
# Q. wap to print sum of squares of n natural numbers
n=int(input("enter the last number"))
i=1
sum=0
while (i <= n):
  sum = sum + i*i
  i =i+1

print("the sum is:", sum)
```

```python
#loop   BREAK STATEMENT
i=1
while (i < 10):
  print(i)
  if (i==5):
    break
  i=i+1
```

```python
# CONTINUE STATEMENT
i=0
while (i < 10):
  i=i+1
  if (i==5):
    continue
  print(i)
```

```python
#BREAK OPRATOR
k1=['a','b','c','d']
k2=['e','f','g','h']
for x in k1:
  if x=='b':
    break
  print(x)
```
